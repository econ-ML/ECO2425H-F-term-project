### A collider

```python
graph = graphviz.Digraph(format='png', engine='neato')

graph.node('T', label='Financial Education', fontsize='10')
graph.node('Z', label='Selection Bias \nFactor', fontsize='10')
graph.node('D', label='Follow Up Survey\n Participation', fontsize='10')
graph.node('Y', label='Student Credit\n Card Usage', fontsize='10')

nodes = ['T', 'Z', 'D', 'Y']
positions = ['0,1!', '2,1!', '0,0!', '2,0!']

edges = ['TD', 'ZD', 'TY', 'ZY']

[graph.node(n, pos=pos) for n, pos in zip(nodes, positions)]

graph.edges(edges)

graph.render('img/dag_structure', view=True)

graph

```
